```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr, spearmanr

# Create data
np.random.seed(42)
#seed generates the same values in the session

# Positive correlation
x_pos = np.arange(1, 11)
y_pos = x_pos + np.random.normal(0, 1, 10) # y increases with x with added noise

# Negative correlation
x_neg = np.arange(1, 11)
y_neg = -x_neg + np.random.normal(0, 1, 10) # y decreases with x with added noise

# No correlation
x_none = np.arange(1, 11)
y_none = np.random.normal(0, 5, 10) # y is randomly distributed regardless of x

# Plot settings
fig, axs = plt.subplots(1, 3, figsize=(18, 5)) # Create a figure and a set of subplots

datasets = [(x_pos, y_pos, "Positive Correlation"),
            (x_neg, y_neg, "Negative Correlation"),
            (x_none, y_none, "No Correlation")]

colors = ['green', 'red', 'blue']

for i, (x, y, title) in enumerate(datasets):
    # Calculate Pearson and Spearman correlation coefficients
    pearson_r, _ = pearsonr(x, y)
    spearman_r, _ = spearmanr(x, y)

    # Plot the data points
    axs[i].scatter(x, y, color=colors[i])
    # Set the title with correlation coefficients
    axs[i].set_title(f"{title}\nPearson r = {pearson_r:.2f}, Spearman  = {spearman_r:.2f}")
    axs[i].set_xlabel("X")
    axs[i].set_ylabel("Y")
    # Add trend line
    m, b = np.polyfit(x, y, 1)
    axs[i].plot(x, m*x + b, color='black', linestyle='--')

# Set the main title for the figure
plt.suptitle("Pearson vs. Spearman Correlation Examples", fontsize=16)
# Adjust layout to prevent overlapping titles and labels
plt.tight_layout(rect=[0, 0, 1, 0.95])
# Display the plot
plt.show()
```

# Task
Perform non-parametric hypothesis tests (Mann-Whitney U, Kruskal-Wallis, Wilcoxon Signed-Rank, and Friedman tests) on the "Titanic: Machine Learning from Disaster" dataset from "https://www.kaggle.com/c/titanic/data" and simulated data. For the Mann-Whitney U test, compare age distributions between survived and not survived passengers. For the Kruskal-Wallis test, test if fare paid differs across the three passenger classes. ### Subtask:
Load the "Titanic: Machine Learning from Disaster" dataset into a pandas DataFrame.


**Reasoning**:
Load the "Titanic: Machine Learning from Disaster" dataset into a pandas DataFrame.



```python
import pandas as pd

df_titanic = pd.read_csv('train.csv') # Assuming 'train.csv' is the filename for the Titanic dataset

display(df_titanic.head())
display(df_titanic.info())
```

## Mann-whitney u test

### Subtask:
Prepare the data by separating ages for 'survived' and 'not survived' passengers, handling missing age values appropriately. Perform the Mann-Whitney U test to compare the age distributions. Interpret the results in the context of the null and alternative hypotheses.


**Reasoning**:
Separate the age data based on survival status, handle missing values by dropping rows with NaN in the 'Age' column, and then perform the Mann-Whitney U test and print the results and interpretation.



```python
from scipy.stats import mannwhitneyu

# Handle missing age values and separate data
df_titanic_cleaned = df_titanic.dropna(subset=['Age'])
age_survived = df_titanic_cleaned[df_titanic_cleaned['Survived'] == 1]['Age']
age_not_survived = df_titanic_cleaned[df_titanic_cleaned['Survived'] == 0]['Age']

# Perform the Mann-Whitney U test
statistic, p_value = mannwhitneyu(age_survived, age_not_survived)

# Print the results
print(f"Mann-Whitney U statistic: {statistic:.2f}")
print(f"P-value: {p_value:.4f}")

# Interpret the results
alpha = 0.05
print("\nInterpretation:")
print("Null Hypothesis (H0): The distribution of age is the same for survived and not survived passengers.")
print("Alternative Hypothesis (H1): The distribution of age is different for survived and not survived passengers.")

if p_value < alpha:
    print(f"Since the p-value ({p_value:.4f}) is less than the significance level ({alpha}), we reject the null hypothesis.")
    print("Conclusion: There is a statistically significant difference in the age distributions between passengers who survived and those who did not survive.")
else:
    print(f"Since the p-value ({p_value:.4f}) is greater than the significance level ({alpha}), we fail to reject the null hypothesis.")
    print("Conclusion: There is no statistically significant difference in the age distributions between passengers who survived and those who did not survive.")
```

## Kruskal-wallis test

### Subtask:
Prepare the data by grouping fare data by passenger class (Pclass), handling missing fare values if any. Perform the Kruskal-Wallis test to see if fare paid differs across the three passenger classes. Interpret the results in the context of the null and alternative hypotheses.


**Reasoning**:
Import the kruskal function, identify unique passenger classes, group fare data by class, perform the Kruskal-Wallis test, and print and interpret the results.



```python
from scipy.stats import kruskal

# Identify unique passenger classes
pclass_groups = df_titanic['Pclass'].unique()

# Create a list to hold fare data for each class
fare_by_class = []

# Group fare data by passenger class
for pclass in pclass_groups:
    # Filter the DataFrame for the current class and get the 'Fare' values
    # Handle potential missing values, though df.info() showed none in the dummy data
    class_fares = df_titanic[df_titanic['Pclass'] == pclass]['Fare'].dropna()
    fare_by_class.append(class_fares)

# Perform the Kruskal-Wallis test
# The kruskal function takes the arrays/lists of data for each group as separate arguments
statistic, p_value = kruskal(*fare_by_class)

# Print the results
print(f"Kruskal-Wallis statistic: {statistic:.2f}")
print(f"P-value: {p_value:.4f}")

# Interpret the results
alpha = 0.05
print("\nInterpretation:")
print("Null Hypothesis (H0): The distribution of fare is the same across the three passenger classes.")
print("Alternative Hypothesis (H1): The distribution of fare is different for at least one pair of passenger classes.")

if p_value < alpha:
    print(f"Since the p-value ({p_value:.4f}) is less than the significance level ({alpha}), we reject the null hypothesis.")
    print("Conclusion: There is a statistically significant difference in the distribution of fare paid across the three passenger classes.")
else:
    print(f"Since the p-value ({p_value:.4f}) is greater than the significance level ({alpha}), we fail to reject the null hypothesis.")
    print("Conclusion: There is no statistically significant difference in the distribution of fare paid across the three passenger classes.")
```

## Wilcoxon signed-rank test

### Subtask:
Create a simulated pre-post scenario by selecting a subset of passengers and simulating 'before' and 'after' fare values with a hypothetical promotion effect. Perform the Wilcoxon Signed-Rank test on the simulated paired data. Interpret the results.


**Reasoning**:
The subtask requires performing a Wilcoxon Signed-Rank test on simulated pre-post fare data. This involves selecting a subset of data, simulating the 'before' and 'after' scenarios, and then applying the Wilcoxon test. The necessary function `wilcoxon` needs to be imported.



```python
from scipy.stats import wilcoxon
import numpy as np

# 2. Select a subset of passengers (e.g., the first 10)
# Ensure the subset is not larger than the actual number of rows in the dummy df
subset_size = min(10, len(df_titanic))
df_subset = df_titanic.head(subset_size).copy()

# 3. Create a simulated 'before' fare series
before_fare = df_subset['Fare']

# 4. Create a simulated 'after' fare series with a hypothetical discount
# Let's simulate a 10% discount plus a small random variation
np.random.seed(42) # for reproducibility of the simulation
discount_factor = 0.90
random_noise = np.random.normal(0, 1, size=subset_size) # Add some noise
after_fare = before_fare * discount_factor + random_noise

# Ensure 'after_fare' does not go below zero, although unlikely with a discount
after_fare[after_fare < 0] = 0

# 5. Perform the Wilcoxon Signed-Rank test
# The test requires paired data, which we have with 'before_fare' and 'after_fare'
# It tests if the median difference between paired samples is zero.
statistic, p_value = wilcoxon(before_fare, after_fare)

# 6. Print the calculated Wilcoxon statistic and the p-value
print(f"Wilcoxon Signed-Rank statistic: {statistic:.2f}")
print(f"P-value: {p_value:.4f}")

# 7. Interpret the results
alpha = 0.05
print("\nInterpretation:")
print("Null Hypothesis (H0): The median difference between paired 'before' and 'after' fare values is zero.")
print("Alternative Hypothesis (H1): The median difference between paired 'before' and 'after' fare values is not zero.")

if p_value < alpha:
    print(f"Since the p-value ({p_value:.4f}) is less than the significance level ({alpha}), we reject the null hypothesis.")
    print("Conclusion: There is a statistically significant difference in the median fare values before and after the hypothetical promotion.")
else:
    print(f"Since the p-value ({p_value:.4f}) is greater than the significance level ({alpha}), we fail to reject the null hypothesis.")
    print("Conclusion: There is no statistically significant difference in the median fare values before and after the hypothetical promotion.")
```

Standardize the chemical property columns as PCA is sensitive to the scale of the features.

```python
# Identify chemical property columns (all except 'type' and 'quality')
chemical_cols = df_wine.columns.difference(['type', 'quality'])

# Instantiate StandardScaler
scaler_wine = StandardScaler()

# Fit and transform the chemical property columns
X_scaled_wine = scaler_wine.fit_transform(df_wine[chemical_cols])

# Display the shape of the scaled data to confirm it was created
print(f"Shape of scaled chemical properties: {X_scaled_wine.shape}")
```

## Perform pca

### Subtask:
Apply PCA to the standardized data.


**Reasoning**:
Apply PCA to the standardized wine data.



```python
# Instantiate a PCA object
pca_wine = PCA()

# Fit the PCA model to the X_scaled_wine data
pca_wine.fit(X_scaled_wine)

# Transform the X_scaled_wine data
X_pca_wine = pca_wine.transform(X_scaled_wine)

# Print the shape of the X_pca_wine
print(f"Shape of PCA-transformed data: {X_pca_wine.shape}")
```

**Reasoning**:
The previous PCA step failed because the standardized data still contains NaN values. I need to handle these missing values before applying PCA. I will use an imputer to fill the missing values with the mean of each column.



```python
from sklearn.impute import SimpleImputer

# Instantiate SimpleImputer with mean strategy
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')

# Fit and transform the scaled data to fill missing values
X_scaled_imputed_wine = imputer.fit_transform(X_scaled_wine)

# Instantiate a PCA object
pca_wine = PCA()

# Fit the PCA model to the imputed scaled data
pca_wine.fit(X_scaled_imputed_wine)

# Transform the imputed scaled data
X_pca_wine = pca_wine.transform(X_scaled_imputed_wine)

# Print the shape of the X_pca_wine
print(f"Shape of PCA-transformed data: {X_pca_wine.shape}")
```

Perform Factor Analysis on the prepared data.


**Reasoning**:
Import the FactorAnalysis class, instantiate it, fit it to the imputed data, and transform the data to get the factors.
