## Perform independent samples t-tests

### Subtask:
Perform independent samples t-tests to compare the means of 'MonthlyIncome' and 'Age' between the two groups.


**Reasoning**:
Perform independent samples t-tests for 'MonthlyIncome' and 'Age' between the two attrition groups and print the results.



```python
from scipy.stats import ttest_ind

monthly_income_yes = df[df['Attrition'] == 'Yes']['MonthlyIncome']
monthly_income_no = df[df['Attrition'] == 'No']['MonthlyIncome']

age_yes = df[df['Attrition'] == 'Yes']['Age']
age_no = df[df['Attrition'] == 'No']['Age']

ttest_monthly_income = ttest_ind(monthly_income_yes, monthly_income_no)
ttest_age = ttest_ind(age_yes, age_no)

print("Monthly Income t-test result:")
print(f"T-statistic: {ttest_monthly_income.statistic}")
print(f"P-value: {ttest_monthly_income.pvalue}")

print("\nAge t-test result:")
print(f"T-statistic: {ttest_age.statistic}")
print(f"P-value: {ttest_age.pvalue}")
```

## Interpret the results

### Subtask:
Based on the p-values from the t-tests, determine if there is a statistically significant difference in monthly income and age between employees who left and those who stayed.


**Reasoning**:
Recall the p-values from the t-tests and compare them to a significance level of 0.05 to determine statistical significance.



```python
alpha = 0.05

monthly_income_pvalue = 7.15e-10
age_pvalue = 8.36e-10

print("Statistical significance test results:")

if monthly_income_pvalue < alpha:
    print(f"Monthly Income: P-value ({monthly_income_pvalue:.2e}) is less than alpha ({alpha}), so there is a statistically significant difference in monthly income between employees who left and those who stayed.")
else:
    print(f"Monthly Income: P-value ({monthly_income_pvalue:.2e}) is greater than or equal to alpha ({alpha}), so there is no statistically significant difference in monthly income between employees who left and those who stayed.")

if age_pvalue < alpha:
    print(f"Age: P-value ({age_pvalue:.2e}) is less than alpha ({alpha}), so there is a statistically significant difference in age between employees who left and those who stayed.")
else:
    print(f"Age: P-value ({age_pvalue:.2e}) is greater than or equal to alpha ({alpha}), so there is no statistically significant difference in age between employees who left and those who stayed.")
```

**Reasoning**:
The visual inspection of the distributions is complete. Now, perform the Shapiro-Wilk test to formally check for normality for both 'MonthlyIncome' and 'Age' for each 'Attrition' group.



```python
from scipy.stats import shapiro

# Shapiro-Wilk test for MonthlyIncome
shapiro_monthly_income_yes = shapiro(df[df['Attrition'] == 'Yes']['MonthlyIncome'])
shapiro_monthly_income_no = shapiro(df[df['Attrition'] == 'No']['MonthlyIncome'])

print("Shapiro-Wilk test for Monthly Income:")
print(f"Attrition 'Yes': Statistic={shapiro_monthly_income_yes.statistic}, P-value={shapiro_monthly_income_yes.pvalue}")
print(f"Attrition 'No': Statistic={shapiro_monthly_income_no.statistic}, P-value={shapiro_monthly_income_no.pvalue}")

# Shapiro-Wilk test for Age
shapiro_age_yes = shapiro(df[df['Attrition'] == 'Yes']['Age'])
shapiro_age_no = shapiro(df[df['Attrition'] == 'No']['Age'])

print("\nShapiro-Wilk test for Age:")
print(f"Attrition 'Yes': Statistic={shapiro_age_yes.statistic}, P-value={shapiro_age_yes.pvalue}")
print(f"Attrition 'No': Statistic={shapiro_age_no.statistic}, P-value={shapiro_age_no.pvalue}")
```

**Reasoning**:
Interpret the results of the Shapiro-Wilk tests to determine if the distributions of 'MonthlyIncome' and 'Age' for each 'Attrition' group are normally distributed based on the p-values.



```python
alpha = 0.05

print("Interpretation of Shapiro-Wilk test results:")

# Monthly Income
print("\nMonthly Income:")
if shapiro_monthly_income_yes.pvalue < alpha:
    print(f"Attrition 'Yes': P-value ({shapiro_monthly_income_yes.pvalue:.2e}) is less than alpha ({alpha}), suggesting the distribution is not normally distributed.")
else:
    print(f"Attrition 'Yes': P-value ({shapiro_monthly_income_yes.pvalue:.2e}) is greater than or equal to alpha ({alpha}), suggesting the distribution is normally distributed.")

if shapiro_monthly_income_no.pvalue < alpha:
    print(f"Attrition 'No': P-value ({shapiro_monthly_income_no.pvalue:.2e}) is less than alpha ({alpha}), suggesting the distribution is not normally distributed.")
else:
    print(f"Attrition 'No': P-value ({shapiro_monthly_income_no.pvalue:.2e}) is greater than or equal to alpha ({alpha}), suggesting the distribution is normally distributed.")

# Age
print("\nAge:")
if shapiro_age_yes.pvalue < alpha:
    print(f"Attrition 'Yes': P-value ({shapiro_age_yes.pvalue:.2e}) is less than alpha ({alpha}), suggesting the distribution is not normally distributed.")
else:
    print(f"Attrition 'Yes': P-value ({shapiro_age_yes.pvalue:.2e}) is greater than or equal to alpha ({alpha}), suggesting the distribution is normally distributed.")

if shapiro_age_no.pvalue < alpha:
    print(f"Attrition 'No': P-value ({shapiro_age_no.pvalue:.2e}) is less than alpha ({alpha}), suggesting the distribution is not normally distributed.")
else:
    print(f"Attrition 'No': P-value ({shapiro_age_no.pvalue:.2e}) is greater than or equal to alpha ({alpha}), suggesting the distribution is normally distributed.")
```

## Perform Mann-Whitney U Test for Monthly Income

### Subtask:
Perform the Mann-Whitney U test to compare 'MonthlyIncome' between employees who left and those who stayed.

**Reasoning**:
Since the Shapiro-Wilk test indicated that 'MonthlyIncome' is not normally distributed for either 'Attrition' group, the non-parametric Mann-Whitney U test is appropriate to compare the distributions of 'MonthlyIncome' between the two groups.

```python
from scipy.stats import mannwhitneyu

# Separate MonthlyIncome for each attrition group
monthly_income_yes = df[df['Attrition'] == 'Yes']['MonthlyIncome']
monthly_income_no = df[df['Attrition'] == 'No']['MonthlyIncome']

# Perform Mann-Whitney U test
mannwhitneyu_monthly_income = mannwhitneyu(monthly_income_yes, monthly_income_no)

print("Mann-Whitney U test result for Monthly Income:")
print(f"U-statistic: {mannwhitneyu_monthly_income.statistic}")
print(f"P-value: {mannwhitneyu_monthly_income.pvalue}")
```

## Interpret the Mann-Whitney U Test Result

### Subtask:
Interpret the results of the Mann-Whitney U test for 'MonthlyIncome'.

**Reasoning**:
Compare the p-value from the Mann-Whitney U test to the significance level (alpha = 0.05) to determine if there is a statistically significant difference in the distribution of 'MonthlyIncome' between the two attrition groups.

```python
alpha = 0.05
mannwhitneyu_monthly_income_pvalue = mannwhitneyu_monthly_income.pvalue

print("Interpretation of Mann-Whitney U test for Monthly Income:")

if mannwhitneyu_monthly_income_pvalue < alpha:
    print(f"P-value ({mannwhitneyu_monthly_income_pvalue:.2e}) is less than alpha ({alpha}), suggesting there is a statistically significant difference in the distribution of Monthly Income between employees who left and those who stayed.")
else:
    print(f"P-value ({mannwhitneyu_monthly_income_pvalue:.2e}) is greater than or equal to alpha ({alpha}), suggesting there is no statistically significant difference in the distribution of Monthly Income between employees who left and those who stayed.")
```

## Conclusion for Monthly Income Hypothesis Test

### Subtask:
Based on the Mann-Whitney U test results, draw a conclusion regarding the hypothesis about 'MonthlyIncome'.

**Reasoning**:
Based on the interpretation of the Mann-Whitney U test results, state whether the null hypothesis (H₀: There is no difference in Monthly Income between employees who left and those who stayed) is rejected or not rejected.

```python
alpha = 0.05
mannwhitneyu_monthly_income_pvalue = mannwhitneyu_monthly_income.pvalue

print("Conclusion for Monthly Income Hypothesis Test:")

if mannwhitneyu_monthly_income_pvalue < alpha:
    print(f"Since the p-value ({mannwhitneyu_monthly_income_pvalue:.2e}) is less than the significance level ({alpha}), we reject the null hypothesis.")
    print("Conclusion: There is a statistically significant difference in Monthly Income between employees who left and those who stayed.")
else:
    print(f"Since the p-value ({mannwhitneyu_monthly_income_pvalue:.2e}) is greater than or equal to the significance level ({alpha}), we fail to reject the null hypothesis.")
    print("Conclusion: There is no statistically significant difference in Monthly Income between employees who left and those who stayed.")
```

## Perform Kruskal-Wallis Test for Job Satisfaction

### Subtask:
Perform the Kruskal-Wallis test to compare 'JobSatisfaction' across different 'JobRole' groups.

**Reasoning**:
Based on the visual inspection and the non-significant result from the ANOVA test in the previous step, the distributions of 'JobSatisfaction' across 'JobRole' groups are likely not normally distributed. Therefore, the non-parametric Kruskal-Wallis test is appropriate to compare the distributions of 'JobSatisfaction' across the different 'JobRole' groups.

```python
from scipy.stats import kruskal

# Create a list of JobSatisfaction for each JobRole group
job_satisfaction_groups_kruskal = [df[df['JobRole'] == role]['JobSatisfaction'] for role in df['JobRole'].unique()]

# Perform Kruskal-Wallis test
kruskal_result = kruskal(*job_satisfaction_groups_kruskal)

print("Kruskal-Wallis test result for Job Satisfaction across Job Roles:")
print(f"H-statistic: {kruskal_result.statistic}")
print(f"P-value: {kruskal_result.pvalue}")
```

## Interpret the Kruskal-Wallis Test Result

### Subtask:
Interpret the results of the Kruskal-Wallis test for 'JobSatisfaction'.

**Reasoning**:
Compare the p-value from the Kruskal-Wallis test to the significance level (alpha = 0.05) to determine if there is a statistically significant difference in the distribution of 'JobSatisfaction' across the different 'JobRole' groups.

```python
alpha = 0.05
kruskal_pvalue = kruskal_result.pvalue

print("Interpretation of Kruskal-Wallis test for Job Satisfaction:")

if kruskal_pvalue < alpha:
    print(f"P-value ({kruskal_pvalue:.3f}) is less than alpha ({alpha}), suggesting there is a statistically significant difference in the distribution of Job Satisfaction across job roles.")
else:
    print(f"P-value ({kruskal_pvalue:.3f}) is greater than or equal to alpha ({alpha}), suggesting there is no statistically significant difference in the distribution of Job Satisfaction across job roles.")
```

## Conclusion for Job Satisfaction Hypothesis Test

### Subtask:
Based on the Kruskal-Wallis test results, draw a conclusion regarding the hypothesis about 'JobSatisfaction' across job roles.

**Reasoning**:
Based on the interpretation of the Kruskal-Wallis test results, state whether the null hypothesis (H₀: The median job satisfaction is the same across all job roles) is rejected or not rejected.

```python
alpha = 0.05
kruskal_pvalue = kruskal_result.pvalue

print("Conclusion for Job Satisfaction Hypothesis Test:")

if kruskal_pvalue < alpha:
    print(f"Since the p-value ({kruskal_pvalue:.3f}) is less than the significance level ({alpha}), we reject the null hypothesis.")
    print("Conclusion: There is a statistically significant difference in Job Satisfaction across job roles.")
else:
    print(f"Since the p-value ({kruskal_pvalue:.3f}) is greater than or equal to the significance level ({alpha}), we fail to reject the null hypothesis.")
    print("Conclusion: There is no statistically significant difference in Job Satisfaction across job roles.")
```

```python
from scipy.stats import f_oneway

# Create a list of Age values for each unique JobRole
job_roles = df['JobRole'].unique()
age_groups = [df[df['JobRole'] == role]['Age'] for role in job_roles]

# Perform one-way ANOVA test
f_statistic_age_anova, p_value_age_anova = f_oneway(*age_groups)

print("One-way ANOVA test result for Age across Job Roles:")
print(f"F-statistic: {f_statistic_age_anova}")
print(f"P-value: {p_value_age_anova}")
```

## Perform kruskal-wallis test for age

### Subtask:
Perform a Kruskal-Wallis test to compare the distributions of 'Age' across different 'JobRole' groups.


**Reasoning**:
Perform the Kruskal-Wallis test to compare the distributions of 'Age' across different 'JobRole' groups as requested by the subtask.



```python
from scipy.stats import kruskal

# Create a list of Age values for each unique JobRole
job_roles = df['JobRole'].unique()
age_groups_kruskal = [df[df['JobRole'] == role]['Age'] for role in job_roles]

# Perform Kruskal-Wallis test
kruskal_result_age = kruskal(*age_groups_kruskal)

print("Kruskal-Wallis test result for Age across Job Roles:")
print(f"H-statistic: {kruskal_result_age.statistic}")
print(f"P-value: {kruskal_result_age.pvalue}")
```
